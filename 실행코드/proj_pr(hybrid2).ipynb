```python

# Colab: DeiT-Small (ViT baseline) training + plots
# (PyTorch AdamW + Warmup→Cosine scheduler 버전)
!pip -q install timm==0.9.16 scikit-learn

import os, random, shutil, numpy as np, matplotlib.pyplot as plt
from glob import glob
from google.colab import drive
import torch, torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import timm
from sklearn.metrics import classification_report, confusion_matrix, f1_score

# 0)Mount & Paths
drive.mount('/content/drive')
BASE = "/content/drive/MyDrive/인공지능 튜플"
SRC  = os.path.join(BASE, "OriginalDataset")             # 원본(클래스 4개)
WORK = "/content/dataset_vit"                            # 분할이 만들어질 위치
OUT  = os.path.join(BASE, "proj_pr", "out"); os.makedirs(OUT, exist_ok=True)

# 1) Stratified split builder (8:1:1)
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)
assert os.path.isdir(SRC), f"원본 폴더 없음: {SRC}"

def ensure(p): os.makedirs(p, exist_ok=True)
def list_images(cls_dir):
    files=[]
    for ext in ('*.png','*.jpg','*.jpeg','*.bmp'):
        files += glob(os.path.join(cls_dir, ext))
    return sorted(files)

if os.path.isdir(WORK): shutil.rmtree(WORK)
classes = sorted([d for d in os.listdir(SRC) if os.path.isdir(os.path.join(SRC,d))])
for split in ('train','val','test'):
    for c in classes: ensure(os.path.join(WORK, split, c))

summary = {}
for c in classes:
    files = list_images(os.path.join(SRC, c))
    random.shuffle(files)
    n = len(files); n_val = n // 10; n_test = n // 10
    val_f  = files[:n_val]; test_f = files[n_val:n_val+n_test]; train_f = files[n_val+n_test:]
    for fp in train_f: shutil.copy2(fp, os.path.join(WORK, 'train', c, os.path.basename(fp)))
    for fp in val_f:   shutil.copy2(fp, os.path.join(WORK, 'val',   c, os.path.basename(fp)))
    for fp in test_f:  shutil.copy2(fp, os.path.join(WORK, 'test',  c, os.path.basename(fp)))
    summary[c] = (len(train_f), len(val_f), len(test_f))

print("[Split] →", WORK)
for c,(tr,v,te) in summary.items():
    print(f"  {c:18s}  train={tr:5d}  val={v:4d}  test={te:4d}")

# 2) Dataloaders
IMG_SIZE = 224
train_tf = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])
eval_tf = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize(256), transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

train_ds = datasets.ImageFolder(os.path.join(WORK,'train'), transform=train_tf)
val_ds   = datasets.ImageFolder(os.path.join(WORK,'val'),   transform=eval_tf)
test_ds  = datasets.ImageFolder(os.path.join(WORK,'test'),  transform=eval_tf)
class_names = train_ds.classes; num_classes = len(class_names)
print("Classes:", class_names)

BATCH, NUM_WORKERS = 64, 2
train_loader = DataLoader(train_ds, batch_size=BATCH, shuffle=True,  num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=BATCH, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

#3) Model / Loss / Optim / Scheduler
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = timm.create_model('deit_small_patch16_224', pretrained=True, num_classes=num_classes).to(device)

# 클래스 불균형 완화용 가중치
train_counts = np.bincount(np.array(train_ds.targets), minlength=num_classes)
w = 1.0 / np.clip(train_counts, 1, None); w = w / w.sum() * num_classes
criterion = nn.CrossEntropyLoss(weight=torch.tensor(w, dtype=torch.float, device=device))

LR, WD, EPOCHS = 3e-4, 5e-2, 30

# --- 여기부터 변경: PyTorch 기본 AdamW + Warmup→Cosine 스케줄러
import torch.optim as optim
from torch.optim.lr_scheduler import SequentialLR, LinearLR, CosineAnnealingLR

optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=WD)
warmup_t = max(1, EPOCHS // 10)
warmup = LinearLR(optimizer, start_factor=1e-6 / LR, total_iters=warmup_t)
cosine = CosineAnnealingLR(optimizer, T_max=EPOCHS - warmup_t, eta_min=LR * 1e-2)
scheduler = SequentialLR(optimizer, [warmup, cosine], milestones=[warmup_t])
# --------------------------------------------------------------------

scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())

# -------- 4) Train / Eval helpers --------
def run_epoch(loader, train=True, need_preds=False):
    model.train() if train else model.eval()
    total_loss, total_correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    for x,y in loader:
        x,y = x.to(device, non_blocking=True), y.to(device, non_blocking=True)
        if train:
            optimizer.zero_grad(set_to_none=True)
            with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
                logits = model(x); loss = criterion(logits, y)
            scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
        else:
            with torch.no_grad():
                logits = model(x); loss = criterion(logits, y)
        total_loss += loss.item() * x.size(0)
        pred = logits.argmax(1)
        total_correct += (pred==y).sum().item()
        total += x.size(0)
        if need_preds:
            all_preds.append(pred.detach().cpu().numpy()); all_labels.append(y.detach().cpu().numpy())
    avg_loss = total_loss/total; acc = total_correct/total
    if need_preds:
        all_preds = np.concatenate(all_preds); all_labels = np.concatenate(all_labels)
        macro_f1 = f1_score(all_labels, all_preds, average="macro")
        return avg_loss, acc, macro_f1, all_preds, all_labels
    return avg_loss, acc

# 5) Train loop (keep best by val macro-F1)
history = {"epoch":[], "train_loss":[], "train_acc":[], "val_loss":[], "val_acc":[], "val_f1":[]}
best_f1, best_path = -1.0, None

for ep in range(1, EPOCHS+1):
    tr_loss, tr_acc = run_epoch(train_loader, train=True)
    val_loss, val_acc, val_f1, _, _ = run_epoch(val_loader, train=False, need_preds=True)
    scheduler.step()  # changed (한 epoch마다 스케줄러 진행)
    history["epoch"].append(ep)
    history["train_loss"].append(tr_loss); history["train_acc"].append(tr_acc)
    history["val_loss"].append(val_loss); history["val_acc"].append(val_acc); history["val_f1"].append(val_f1)
    print(f"[{ep:02d}/{EPOCHS}] train_loss={tr_loss:.4f} acc={tr_acc:.4f} | val_loss={val_loss:.4f} acc={val_acc:.4f} f1={val_f1:.4f}")
    if val_f1 > best_f1:
        best_f1 = val_f1
        best_path = os.path.join(OUT, f"best_deitS_f1{best_f1:.4f}.pt")
        torch.save({"model": model.state_dict(), "classes": class_names}, best_path)

print("Best checkpoint:", best_path)

# 6) Plots
plt.figure(); plt.plot(history["epoch"], history["train_loss"], label="train_loss")
plt.plot(history["epoch"], history["val_loss"], label="val_loss")
plt.xlabel("epoch"); plt.ylabel("loss"); plt.title("Loss vs Epoch"); plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig(os.path.join(OUT, "vit_loss_curve.png"), dpi=200); plt.show()

plt.figure(); plt.plot(history["epoch"], history["train_acc"], label="train_acc")
plt.plot(history["epoch"], history["val_acc"], label="val_acc")
plt.xlabel("epoch"); plt.ylabel("accuracy"); plt.title("Accuracy vs Epoch"); plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig(os.path.join(OUT, "vit_accuracy_curve.png"), dpi=200); plt.show()

#7) Test evaluation
from sklearn.metrics import classification_report, confusion_matrix
if best_path:
    ckpt = torch.load(best_path, map_location=device); model.load_state_dict(ckpt["model"])

te_loss, te_acc, te_f1, te_preds, te_labels = run_epoch(test_loader, train=False, need_preds=True)
print(f"\n[TEST] loss={te_loss:.4f}  acc={te_acc:.4f}  macroF1={te_f1:.4f}")
report = classification_report(te_labels, te_preds, target_names=class_names, digits=4)
print("\n[TEST classification report]\n", report)
with open(os.path.join(OUT, "vit_test_report.txt"), "w") as f: f.write(report)

cm = confusion_matrix(te_labels, te_preds)
plt.figure(figsize=(5.6,4.8)); plt.imshow(cm, cmap='Blues'); plt.title('Confusion Matrix (Test, DeiT-S)')
plt.colorbar(); ticks = np.arange(num_classes)
plt.xticks(ticks, class_names, rotation=45, ha='right'); plt.yticks(ticks, class_names)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        v = cm[i,j]; c = 'white' if v>cm.max()/2 else 'black'
        plt.text(j, i, str(v), ha='center', va='center', color=c, fontsize=9)
plt.ylabel('True'); plt.xlabel('Pred'); plt.tight_layout()
plt.savefig(os.path.join(OUT, "vit_confusion_matrix_test.png"), dpi=220); plt.show()

print("\nArtifacts saved in:", OUT)
```

```python
# CNN-ViT Hybrid (ResNet50 backbone -> ViT student)
# - Balanced training via sample_from_datasets + strong aug (~10x exposure)
# - Label smoothing, AdamW + Cosine decay
# - Plots + reports (macro-F1)
import os, glob, random, shutil, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, confusion_matrix, f1_score

# 0) Paths & split
from google.colab import drive
drive.mount('/content/drive')

BASE = '/content/drive/MyDrive/인공지능 튜플'
SRC  = f'{BASE}/OriginalDataset'           # 클래스 폴더 4개
SPLIT= '/content/dataset_hybrid'           # 계층 분할 대상
OUT  = f'{BASE}/proj_pr/out'; os.makedirs(OUT, exist_ok=True)

SEED=123
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

def ensure(p): os.makedirs(p, exist_ok=True)
def list_images(cls_dir):
    files=[]
    for ext in ('*.png','*.jpg','*.jpeg','*.bmp'):
        files += glob.glob(os.path.join(cls_dir, ext))
    return files

if os.path.isdir(SPLIT): shutil.rmtree(SPLIT)
classes = sorted([d for d in os.listdir(SRC) if os.path.isdir(os.path.join(SRC,d))])
for sp in ('train','val','test'):
    for c in classes: ensure(os.path.join(SPLIT, sp, c))

summary={}
for c in classes:
    files = list_images(os.path.join(SRC,c)); random.shuffle(files)
    n=len(files); n_val=n//10; n_test=n//10
    val_f=files[:n_val]; test_f=files[n_val:n_val+n_test]; train_f=files[n_val+n_test:]
    for fp in train_f: shutil.copy2(fp, os.path.join(SPLIT,'train',c,os.path.basename(fp)))
    for fp in val_f:   shutil.copy2(fp, os.path.join(SPLIT,'val',c,os.path.basename(fp)))
    for fp in test_f:  shutil.copy2(fp, os.path.join(SPLIT,'test',c,os.path.basename(fp)))
    summary[c]=(len(train_f),len(val_f),len(test_f))
print("[Split]")
for c,(tr,v,te) in summary.items():
    print(f"  {c:18s}  train={tr:5d}  val={v:4d}  test={te:4d}")

# 1) Datasets
IMG_SIZE=(288,288); BATCH=32
AUTOTUNE=tf.data.AUTOTUNE

# 강한 증강(학습) / 약한 증강(평가)
def aug_strong(x):
    x = tf.image.random_flip_left_right(x)
    x = tf.image.random_brightness(x, 0.08)
    x = tf.image.random_contrast(x, 0.9, 1.1)
    x = tf.image.random_saturation(x, 0.9, 1.1)
    # 랜덤 리사이즈 크롭(zoom/shift 효과)
    crop_scale = tf.random.uniform([], 0.8, 1.0)
    new_h = tf.cast(IMG_SIZE[0]*crop_scale, tf.int32)
    new_w = tf.cast(IMG_SIZE[1]*crop_scale, tf.int32)
    x = tf.image.resize(x, (new_h, new_w))
    x = tf.image.resize_with_pad(x, IMG_SIZE[0], IMG_SIZE[1])
    return x

def load_image(path, label, train=False):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, IMG_SIZE)
    if train: img = aug_strong(img)
    x = img/255.0
    return x, label

# per-class file lists (train)
per_class_files = {}
for i,c in enumerate(classes):
    per_class_files[c] = sorted(glob.glob(os.path.join(SPLIT,'train',c,'*')))

# per-class datasets (repeat + strong aug)
per_class_ds=[]
class_counts=[]
for i,c in enumerate(classes):
    files = tf.constant(per_class_files[c])
    labels= tf.fill([tf.shape(files)[0]], i)
    ds = tf.data.Dataset.from_tensor_slices((files, labels)).shuffle(4096, seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(lambda p,l: load_image(p,l,train=True), num_parallel_calls=AUTOTUNE).repeat()
    per_class_ds.append(ds)
    class_counts.append(len(per_class_files[c]))

C=len(classes)
# 균형 샘플러: 각 클래스 동일 확률로 샘플
train_ds = tf.data.Dataset.sample_from_datasets(per_class_ds, weights=[1.0/C]*C, stop_on_empty_dataset=False)
# one-hot 라벨 + 배치
train_ds = train_ds.map(lambda x,y: (x, tf.one_hot(y, C)), num_parallel_calls=AUTOTUNE)
train_ds = train_ds.batch(BATCH).prefetch(AUTOTUNE)

# val/test (약한 증강, one-hot)
val_raw = tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT,'val'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)
test_raw= tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT,'test'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)

# 클래스 순서 확인(폴더명 알파벳 기준)
print("Class order:", val_raw.class_names)
assert val_raw.class_names==classes, "폴더/클래스 순서가 train과 다릅니다."

val_ds  = val_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, C))).prefetch(AUTOTUNE)
test_ds = test_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, C))).prefetch(AUTOTUNE)

# 2) ViT blocks
# --------------------
# 안전한 2D Positional Embedding
# --------------------
class PositionalEmbedding2D(layers.Layer):
    def build(self, input_shape):
        H = input_shape[1]; W = input_shape[2]; D = input_shape[3]
        if H is None or W is None or D is None:
            raise ValueError(f"PositionalEmbedding2D needs static H,W,D, got {input_shape}")
        self.pos = self.add_weight(
            name="pos2d",
            shape=(H, W, D),
            initializer=keras.initializers.RandomNormal(stddev=0.02),
            trainable=True,
        )
    def call(self, x):  # x: (B,H,W,D)
        return x + self.pos

class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads=4, mlp_ratio=4.0, drop=0.1):
        super().__init__()
        self.n1 = layers.LayerNormalization(epsilon=1e-6)
        self.mha= layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim//num_heads, dropout=drop)
        self.d1 = layers.Dropout(drop)
        self.n2 = layers.LayerNormalization(epsilon=1e-6)
        self.mlp= keras.Sequential([
            layers.Dense(int(embed_dim*mlp_ratio), activation=tf.nn.gelu),
            layers.Dropout(drop),
            layers.Dense(embed_dim),
            layers.Dropout(drop),
        ])
    def call(self, x, training=False):
        h = self.mha(self.n1(x), self.n1(x), training=training)
        x = x + self.d1(h, training=training)
        x = x + self.mlp(self.n2(x), training=training)
        return x

def build_hybrid(input_shape, num_classes, embed_dim=256, vit_blocks=4, heads=4, drop=0.1):
    inp = layers.Input(input_shape)
    x = keras.applications.resnet.preprocess_input(inp*255.0)  # inputs in [0,1] → [0,255]
    backbone = keras.applications.ResNet50(include_top=False, weights='imagenet')
    feat = backbone(x, training=True)     # (B, 9, 9, 2048) for 288x288

    # Conv patch embedding
    tok = layers.Conv2D(embed_dim, kernel_size=2, strides=2, padding='same')(feat)  # → (B, 5, 5, 256)
    tok = layers.BatchNormalization()(tok)
    tok = layers.Activation('gelu')(tok)
    tok = layers.Dropout(drop)(tok)

    # 2D learnable position, then flatten to tokens
    tok = PositionalEmbedding2D()(tok)              # add (H,W,D) pos
    H, W = tok.shape[1], tok.shape[2]
    tokens = layers.Reshape((H * W, embed_dim))(tok)  # T=H*W is static now
    z = layers.Dropout(drop)(tokens)

    for _ in range(vit_blocks):
        z = TransformerEncoder(embed_dim=embed_dim, num_heads=heads, mlp_ratio=4.0, drop=drop)(z)

    z = layers.GlobalAveragePooling1D()(z)
    z = layers.Dropout(0.5)(z)
    z = layers.Dense(512, activation='gelu')(z)     # 비선형성 강조
    z = layers.Dropout(0.4)(z)
    out = layers.Dense(num_classes, activation='softmax')(z)

    model = keras.Model(inp, out, name='ResNet50_ViT_hybrid')

    # BN 고정(안정화)
    for l in backbone.layers:
        if isinstance(l, layers.BatchNormalization): l.trainable = False

    return model

model = build_hybrid(IMG_SIZE+(3,), C, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
model.summary()

# 4) Compile
steps_per_epoch = max(200, (sum([n for n,_,_ in summary.values()])//BATCH)*2)  # 균형샘플 + 강한 증강 → 충분한 노출
opt = keras.optimizers.AdamW(learning_rate=1e-4, weight_decay=1e-4)
model.compile(
    optimizer=opt,
    loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
    metrics=['accuracy']
)

ckpt = keras.callbacks.ModelCheckpoint(f'{OUT}/best_hybrid.h5', monitor='val_accuracy', mode='max',
                                       save_best_only=True, verbose=1)
early= keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1)
plate= keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6, verbose=1)

EPOCHS=50
hist = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    steps_per_epoch=steps_per_epoch,
    callbacks=[ckpt, plate, early],
    verbose=1
)

# 5) Plots
epochs = range(1, len(hist.history['loss'])+1)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(epochs, hist.history['accuracy'], label='Train'); plt.plot(epochs, hist.history['val_accuracy'], label='Val')
plt.title('Accuracy'); plt.xlabel('Epoch'); plt.legend(); plt.grid(True)
plt.subplot(1,2,2)
plt.plot(epochs, hist.history['loss'], label='Train'); plt.plot(epochs, hist.history['val_loss'], label='Val')
plt.title('Loss'); plt.xlabel('Epoch'); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.savefig(f'{OUT}/hybrid_training_curves.png', dpi=200); plt.show()

# 6) Reports (Val/Test, macro-F1)
def eval_report(ds, classes, tag):
    # predict
    y_true, y_pred = [], []
    for x, y in ds:
        p = model.predict(x, verbose=0)
        y_true.extend(np.argmax(y.numpy(), axis=1))
        y_pred.extend(np.argmax(p, axis=1))
    rep = classification_report(y_true, y_pred, target_names=classes, digits=4)
    print(f"\n[{tag} report]\n", rep)
    with open(f'{OUT}/hybrid_{tag}_report.txt','w') as f: f.write(rep)
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5.6,4.8))
    plt.imshow(cm, cmap='Blues'); plt.title(f'Confusion Matrix ({tag})'); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    for i in range(len(classes)):
        for j in range(len(classes)):
            v=cm[i,j]; c='white' if v>cm.max()/2 else 'black'
            plt.text(j,i,str(v),ha='center',va='center',color=c,fontsize=9)
    plt.tight_layout(); plt.savefig(f'{OUT}/hybrid_confmat_{tag}.png', dpi=220); plt.show()
    return f1_score(y_true, y_pred, average='macro')

val_f1 = eval_report(val_ds, classes, 'val')
test_f1= eval_report(test_ds, classes, 'test')
print(f"[Summary] Macro-F1  val={val_f1:.4f}  test={test_f1:.4f}\nSaved to: {OUT}")
```

```python
# ============================================================
# CNN–ViT Hybrid (ResNet50 -> ViT) | Rollout + TTA (with robust CKPT loader)
# - CKPT 우선순위/형식 감지 로더 (*.weights.h5 > .keras(zip) > .h5)
# - Validation/Test 로더 재구성(증강 OFF)
# - Attention Rollout 시각화 (PNG 저장, 안전장치/간단 마스크 포함)
# - TTA 정책별 성능 비교 (CSV 저장)
# 저장 경로: /content/drive/MyDrive/인공지능 튜플/proj_pr/out/
# ============================================================
import os, glob, numpy as np, matplotlib.pyplot as plt, time, zipfile
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score, roc_auc_score, average_precision_score

# -------------------
# 0) 경로/상수
# -------------------
from google.colab import drive
drive.mount('/content/drive')

BASE  = '/content/drive/MyDrive/인공지능 튜플'
SPLIT = '/content/dataset_hybrid'   # 이전 학습 때 만든 split 경로(미리 존재해야 함)
OUT   = f'{BASE}/proj_pr/out'
os.makedirs(OUT, exist_ok=True)

IMG_SIZE = (288, 288)
BATCH = 32
SEED = 123
NUM_CLASSES = 4

tf.keras.utils.set_random_seed(SEED)

# -------------------
# 1) 커스텀 레이어/모델 정의 (직렬화/역직렬화 가능)
# -------------------
class PositionalEmbedding2D(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.pos = None
    def build(self, input_shape):
        H, W, D = input_shape[1], input_shape[2], input_shape[3]
        if None in (H, W, D):
            raise ValueError(f"PositionalEmbedding2D needs static H,W,D, got {input_shape}")
        self.pos = self.add_weight(
            name="pos2d",
            shape=(H, W, D),
            initializer=keras.initializers.RandomNormal(stddev=0.02),
            trainable=True,
        )
    def call(self, x):  # x: (B,H,W,D)
        return x + self.pos
    def get_config(self):
        cfg = super().get_config()
        return cfg

class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads=4, mlp_ratio=4.0, drop=0.1, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.mlp_ratio = mlp_ratio
        self.drop = drop
        self.n1 = layers.LayerNormalization(epsilon=1e-6)
        self.mha= layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim//num_heads, dropout=drop)
        self.d1 = layers.Dropout(drop)
        self.n2 = layers.LayerNormalization(epsilon=1e-6)
        self.mlp= keras.Sequential([
            layers.Dense(int(embed_dim*mlp_ratio), activation=tf.nn.gelu),
            layers.Dropout(drop),
            layers.Dense(embed_dim),
            layers.Dropout(drop),
        ])
    def call(self, x, training=False):
        h = self.mha(self.n1(x), self.n1(x), training=training)
        x = x + self.d1(h, training=training)
        x = x + self.mlp(self.n2(x), training=training)
        return x
    def get_config(self):
        cfg = super().get_config()
        cfg.update(dict(embed_dim=self.embed_dim, num_heads=self.num_heads,
                        mlp_ratio=self.mlp_ratio, drop=self.drop))
        return cfg

def build_hybrid(input_shape, num_classes, embed_dim=256, vit_blocks=4, heads=4, drop=0.1):
    inp = layers.Input(input_shape)
    x = keras.applications.resnet.preprocess_input(inp*255.0)
    backbone = keras.applications.ResNet50(include_top=False, weights='imagenet', name='resnet50')
    feat = backbone(x, training=False)  # 추론/로드 시 안정화

    tok = layers.Conv2D(embed_dim, 2, strides=2, padding='same', name='patch_embed_conv')(feat)
    tok = layers.BatchNormalization(name='patch_embed_bn')(tok)
    tok = layers.Activation('gelu', name='patch_embed_gelu')(tok)
    tok = layers.Dropout(drop, name='patch_embed_dropout')(tok)

    tok = PositionalEmbedding2D(name='pos2d')(tok)
    H, W = tok.shape[1], tok.shape[2]
    tokens = layers.Reshape((H*W, embed_dim), name='tokens')(tok)
    z = layers.Dropout(drop, name='tokens_dropout')(tokens)

    for i in range(vit_blocks):
        z = TransformerEncoder(embed_dim=embed_dim, num_heads=heads, mlp_ratio=4.0, drop=drop, name=f'encoder_{i}')(z)

    z = layers.GlobalAveragePooling1D(name='gap')(z)
    z = layers.Dropout(0.5, name='head_drop1')(z)
    z = layers.Dense(512, activation='gelu', name='head_dense')(z)
    z = layers.Dropout(0.4, name='head_drop2')(z)
    out = layers.Dense(num_classes, activation='softmax', name='classifier')(z)

    model = keras.Model(inp, out, name='ResNet50_ViT_hybrid')
    for l in backbone.layers:
        if isinstance(l, layers.BatchNormalization): l.trainable = False
    return model

# -------------------
# 2) CKPT 선택/로딩 (우선순위 + 형식 감지)
# -------------------
def pick_ckpt(out_dir):
    # 1순위: *.weights.h5 (가중치만)  2순위: 정상 .keras(zip)  3순위: 기타 .h5
    w = sorted(glob.glob(os.path.join(out_dir, '*.weights.h5')), key=os.path.getmtime, reverse=True)
    k = sorted(glob.glob(os.path.join(out_dir, '*.keras')), key=os.path.getmtime, reverse=True)
    h = sorted([p for p in glob.glob(os.path.join(out_dir, '*.h5')) if not p.endswith('.weights.h5')], key=os.path.getmtime, reverse=True)
    cand = (w + k + h)
    print("[CKPT candidates]")
    for p in cand[:8]:
        sz = os.path.getsize(p) if os.path.exists(p) else -1
        print(" -", os.path.basename(p), f"({sz} bytes)")
    return cand[0] if cand else None

def is_valid_keras_zip(path):
    try:
        return zipfile.is_zipfile(path)
    except:
        return False

CKPT = pick_ckpt(OUT)
assert CKPT is not None, f"체크포인트 없음: {OUT}"

roll_model = None

if CKPT.endswith('.weights.h5'):
    print("[i] Detected weights-only HDF5 → build same arch then load_weights")
    tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
    tmp.load_weights(CKPT)
    roll_model = tmp
    print("[✓] load_weights 성공:", os.path.basename(CKPT))

elif CKPT.endswith('.keras'):
    if is_valid_keras_zip(CKPT):
        print("[i] Detected valid .keras zip → load_model")
        roll_model = keras.models.load_model(
            CKPT, compile=False,
            custom_objects={'PositionalEmbedding2D': PositionalEmbedding2D,
                            'TransformerEncoder': TransformerEncoder}
        )
        print("[✓] load_model(.keras) 성공:", os.path.basename(CKPT))
    else:
        print("[!] .keras 파일이 zip 포맷이 아닙니다. 가중치 파일로 간주해 로딩을 시도합니다.")
        try:
            tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
            tmp.load_weights(CKPT)  # 실제 HDF5라면 로드됨
            roll_model = tmp
            print("[✓] load_weights(.keras명) 성공:", os.path.basename(CKPT))
        except Exception as e:
            raise RuntimeError(
                f"'{os.path.basename(CKPT)}'는 유효한 .keras zip도, HDF5 가중치도 아닙니다. "
                "→ *.weights.h5 또는 정상 .keras 파일을 사용하세요."
            ) from e
else:  # 기타 .h5 (전체모델 or 가중치)
    print("[i] .h5 감지 → 우선 load_model 시도, 실패 시 weights로 재시도")
    try:
        roll_model = keras.models.load_model(
            CKPT, compile=False,
            custom_objects={'PositionalEmbedding2D': PositionalEmbedding2D,
                            'TransformerEncoder': TransformerEncoder}
        )
        print("[✓] load_model(.h5) 성공:", os.path.basename(CKPT))
    except Exception as e:
        print("[!] load_model(.h5) 실패 → load_weights 시도:", e)
        tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
        tmp.load_weights(CKPT)
        roll_model = tmp
        print("[✓] load_weights(.h5) 성공:", os.path.basename(CKPT))

# -------------------
# 3) Val/Test 로더 (증강 없이, [0,1] 정규화)
# -------------------
val_raw = tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT, 'val'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)
test_raw = tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT, 'test'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)

classes = val_raw.class_names
val_ds  = val_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, NUM_CLASSES))).prefetch(tf.data.AUTOTUNE)
test_ds = test_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, NUM_CLASSES))).prefetch(tf.data.AUTOTUNE)
print("Classes:", classes)

# -------------------
# 4) Attention Rollout 준비 (레이어 안전 추출)
# -------------------
def get_layer_safe(model, name=None, type_=None, start_after=None):
    if name is not None:
        try:
            return model.get_layer(name)
        except:
            pass
    if type_ is not None:
        layers_ = model.layers
        if start_after is not None:
            layers_ = layers_[model.layers.index(start_after)+1:]
        for l in layers_:
            if isinstance(l, type_): return l
    raise ValueError(f"Layer not found: name={name}, type={type_}")

L = roll_model.layers
resnet = get_layer_safe(roll_model, name='resnet50')
conv_pe = get_layer_safe(roll_model, name='patch_embed_conv', type_=layers.Conv2D, start_after=resnet)
bn_pe   = get_layer_safe(roll_model, name='patch_embed_bn',   type_=layers.BatchNormalization)
gelu_pe = get_layer_safe(roll_model, name='patch_embed_gelu')
drop0   = get_layer_safe(roll_model, name='patch_embed_dropout', type_=layers.Dropout)
pos2d   = get_layer_safe(roll_model, name='pos2d', type_=PositionalEmbedding2D)
reshape = get_layer_safe(roll_model, name='tokens', type_=layers.Reshape)
drop_tk = get_layer_safe(roll_model, name='tokens_dropout', type_=layers.Dropout)
encoders = [l for l in L if isinstance(l, TransformerEncoder)]
assert encoders, "TransformerEncoder 레이어를 찾을 수 없습니다."

@tf.function(jit_compile=False)
def tokens_from_image(x):  # x: [0,1]
    x = keras.applications.resnet.preprocess_input(x*255.0)
    f = resnet(x, training=False)
    t = conv_pe(f, training=False)
    t = bn_pe(t, training=False)
    t = gelu_pe(t)
    t = drop0(t, training=False)
    t = pos2d(t, training=False)
    H = tf.shape(t)[1]; W = tf.shape(t)[2]
    z = reshape(t); z = drop_tk(z, training=False)  # (B,T,D)
    return z, H, W

# -------------------
# 4-1) 안전장치 & 간단 뇌 마스크 유틸
# -------------------
def _assert_row_stochastic(A, atol=1e-4):
    row_sum = tf.reduce_sum(A, axis=-1)  # sum over keys
    tf.debugging.assert_near(row_sum, tf.ones_like(row_sum), atol=atol,
                             message="Attention rows must sum to 1 (row-stochastic).")

def _assert_grid_tokens(H, W, T):
    tf.debugging.assert_equal(tf.math.multiply(H, W), T,
                              message="Grid H*W must equal #tokens (reshape mismatch).")

def brain_mask(x, thr=0.03):
    """ x: [0,1] NHWC. MRI 배경 억제를 위한 매우 간단한 마스크 """
    g = tf.image.rgb_to_grayscale(x)
    m = tf.cast(g > thr, tf.float32)  # 배경(거의 0) 제거
    return m

# -------------------
# 4-2) Attention Rollout (개선 버전)
# -------------------
def attention_rollout(
    images, save_prefix, max_images=8, cmap='jet',
    head_fuse='mean',            # 'mean' 또는 'max'
    use_residual=True,           # A -> (A+I) 정규화
    apply_mask=True, mask_thr=0.03
):
    images_tf = tf.convert_to_tensor(images, dtype=tf.float32) if isinstance(images, np.ndarray) else images

    # 토큰/그리드
    z, H, W = tokens_from_image(images_tf)       # z: (B, T, D)
    B = tf.shape(z)[0]; T = tf.shape(z)[1]
    _assert_grid_tokens(H, W, T)

    # 레이어별 attention 누적
    R = tf.eye(T, batch_shape=[B])               # (B, T, T)
    cur = z
    for enc in encoders:
        q = enc.n1(cur)
        attn_out, scores = enc.mha(q, q, return_attention_scores=True, training=False)  # (B,H,T,T)
        # Head fuse
        if head_fuse == 'max':
            A = tf.reduce_max(scores, axis=1)    # (B, T, T)
        else:
            A = tf.reduce_mean(scores, axis=1)   # (B, T, T)

        _assert_row_stochastic(A)

        # Residual 포함 및 행 정규화
        if use_residual:
            I = tf.eye(T, batch_shape=[B])
            A = A + I
        A = A / tf.reduce_sum(A, axis=-1, keepdims=True)

        # 방향 고정: R = R @ A
        R = tf.linalg.matmul(R, A)

        # 다음 블록의 순전파
        cur = cur + enc.d1(attn_out, training=False)
        h = enc.n2(cur); h = enc.mlp(h, training=False)
        cur = cur + h

    # GAP 분류에 대응: query(행) 평균 → 각 key 토큰의 기여
    token_imp = tf.reduce_mean(R, axis=1)        # (B, T)
    token_imp = tf.reshape(token_imp, (-1, H, W, 1))
    token_imp = tf.image.resize(token_imp, tf.shape(images_tf)[1:3], method='bilinear')

    # 간단 뇌 마스크 적용(옵션)
    if apply_mask:
        m = brain_mask(images_tf, thr=mask_thr)  # (B,H,W,1)
        token_imp = token_imp * m

    # per-image 정규화
    tmin = tf.reduce_min(token_imp, axis=[1,2,3], keepdims=True)
    tmax = tf.reduce_max(token_imp, axis=[1,2,3], keepdims=True)
    token_imp = (token_imp - tmin) / (tmax - tmin + 1e-8)

    # 저장
    token_imp_np = token_imp.numpy()
    n = min(token_imp_np.shape[0], max_images)
    os.makedirs(f"{OUT}/rollout", exist_ok=True)
    base_img = images if isinstance(images, np.ndarray) else images_tf.numpy()
    for i in range(n):
        img_np   = base_img[i]
        heat_np  = token_imp_np[i, ..., 0]
        plt.figure(figsize=(4,4)); plt.axis('off')
        plt.imshow(img_np); plt.imshow(heat_np, cmap=cmap, alpha=0.45)
        plt.tight_layout()
        plt.savefig(f"{OUT}/rollout/{save_prefix}_{i:02d}.png", dpi=180)
        plt.close()

# -------------------
# 4-3) (선택) Gradient-weighted Rollout
# -------------------
def attention_rollout_grad(
    images, save_prefix, max_images=8, head_fuse='mean',
    apply_mask=True, mask_thr=0.03
):
    images_tf = tf.convert_to_tensor(images, dtype=tf.float32) if isinstance(images, np.ndarray) else images
    z, H, W = tokens_from_image(images_tf)
    B = tf.shape(z)[0]; T = tf.shape(z)[1]
    _assert_grid_tokens(H, W, T)

    with tf.GradientTape(persistent=True) as tape:
        tape.watch(z)
        cur = z
        A_list = []
        for enc in encoders:
            q = enc.n1(cur)
            attn_out, scores = enc.mha(q, q, return_attention_scores=True, training=False)  # (B,H,T,T)
            if head_fuse == 'max':
                A = tf.reduce_max(scores, axis=1)
            else:
                A = tf.reduce_mean(scores, axis=1)
            _assert_row_stochastic(A)
            A_list.append(A)

            cur = cur + enc.d1(attn_out, training=False)
            h = enc.n2(cur); h = enc.mlp(h, training=False)
            cur = cur + h

        # 모델 헤드 일부 재사용해 확률 계산
        gap = tf.reduce_mean(cur, axis=1)                        # (B,D)
        h1  = roll_model.get_layer('head_drop1')(gap, training=False)
        h2  = roll_model.get_layer('head_dense')(h1)             # gelu 포함
        h3  = roll_model.get_layer('head_drop2')(h2, training=False)
        probs = roll_model.get_layer('classifier')(h3)           # softmax
        top = tf.argmax(probs, axis=1)
        top_prob = tf.gather_nd(probs, tf.stack([tf.range(B), top], axis=1))  # (B,)

    R = tf.eye(T, batch_shape=[B])
    for A in A_list:
        g = tape.gradient(top_prob, A)          # (B,T,T)
        g = tf.nn.relu(g)                       # 양의 기여만
        A_g = A * g
        I = tf.eye(T, batch_shape=[B])
        A_g = (A_g + I)
        A_g = A_g / tf.reduce_sum(A_g, axis=-1, keepdims=True)
        R = tf.linalg.matmul(R, A_g)

    token_imp = tf.reduce_mean(R, axis=1)       # (B,T)
    token_imp = tf.reshape(token_imp, (-1, H, W, 1))
    token_imp = tf.image.resize(token_imp, tf.shape(images_tf)[1:3], method='bilinear')

    if apply_mask:
        m = brain_mask(images_tf, thr=mask_thr)
        token_imp = token_imp * m

    tmin = tf.reduce_min(token_imp, axis=[1,2,3], keepdims=True)
    tmax = tf.reduce_max(token_imp, axis=[1,2,3], keepdims=True)
    token_imp = (token_imp - tmin) / (tmax - tmin + 1e-8)

    token_imp_np = token_imp.numpy()
    n = min(token_imp_np.shape[0], max_images)
    os.makedirs(f"{OUT}/rollout", exist_ok=True)
    base_img = images if isinstance(images, np.ndarray) else images_tf.numpy()
    for i in range(n):
        img_np   = base_img[i]
        heat_np  = token_imp_np[i, ..., 0]
        plt.figure(figsize=(4,4)); plt.axis('off')
        plt.imshow(img_np); plt.imshow(heat_np, cmap='jet', alpha=0.45)
        plt.tight_layout()
        plt.savefig(f"{OUT}/rollout/{save_prefix}_grad_{i:02d}.png", dpi=180)
        plt.close()

# -------------------
# 5) 예시: validation 배치 8장 시각화
# -------------------
sample_batch = next(iter(val_ds.unbatch().batch(8)))[0].numpy()
attention_rollout(sample_batch, save_prefix="val", max_images=8, head_fuse='mean', apply_mask=True, mask_thr=0.03)
attention_rollout_grad(sample_batch, save_prefix="val", max_images=8, head_fuse='mean', apply_mask=True, mask_thr=0.03)
print("Saved rollout overlays →", f"{OUT}/rollout")

# -------------------
# 6) TTA 성능 비교 (Test)
# -------------------
def five_crops(x, pad=16):
    H, W = x.shape[1], x.shape[2]
    x_pad = tf.pad(x, [[0,0],[pad,pad],[pad,pad],[0,0]], mode='REFLECT')
    crops = []
    crops.append(tf.image.crop_to_bounding_box(x_pad, 0, 0,     H, W))       # TL
    crops.append(tf.image.crop_to_bounding_box(x_pad, 0, 2*pad, H, W))       # TR
    crops.append(tf.image.crop_to_bounding_box(x_pad, 2*pad, 0, H, W))       # BL
    crops.append(tf.image.crop_to_bounding_box(x_pad, 2*pad, 2*pad, H, W))   # BR
    crops.append(tf.image.crop_to_bounding_box(x_pad, pad, pad, H, W))       # Center
    return crops

def predict_tta(batch_x, policy, mdl):
    views = []
    if policy == 'none':
        views = [batch_x]
    elif policy == 'hflip':
        views = [batch_x, tf.image.flip_left_right(batch_x)]
    elif policy == '5crop':
        views = five_crops(batch_x)
    elif policy == 'hflip+5crop':
        base = five_crops(batch_x); views = base + [tf.image.flip_left_right(c) for c in base]
    else:
        raise ValueError("unknown TTA policy")
    probs = []
    for v in views:
        probs.append(mdl.predict(v, verbose=0))   # softmax 확률
    probs = np.stack(probs, axis=0).mean(axis=0)  # 평균 앙상블
    return probs

def eval_ds_tta(ds, classes, policy, mdl):
    y_true, y_pred, probs = [], [], []
    n_img = 0; t0 = time.time()
    for x, y in ds:
        p = predict_tta(x, policy, mdl)
        probs.append(p)
        y_true.extend(np.argmax(y.numpy(), axis=1))
        y_pred.extend(np.argmax(p, axis=1))
        n_img += x.shape[0]
    dt = time.time() - t0

    probs = np.concatenate(probs, 0)
    y_true = np.array(y_true); y_pred = np.array(y_pred)
    acc = accuracy_score(y_true, y_pred)
    mf1 = f1_score(y_true, y_pred, average='macro')
    auc = roc_auc_score(y_true, probs, multi_class='ovr', average='macro')
    ms_img = dt / n_img * 1000.0
    cm = confusion_matrix(y_true, y_pred)

    print(f"[TTA:{policy:11s}] acc={acc:.4f}  macroF1={mf1:.4f}  ROC-AUC={auc:.4f}  {ms_img:.1f} ms/img")
    return acc, mf1, auc, ms_img, cm

policies = ['none', 'hflip', '5crop', 'hflip+5crop']
rows = []
cms  = {}
for pol in policies:
    acc, mf1, auc, ms, cm = eval_ds_tta(test_ds, classes, pol, roll_model)
    rows.append([pol, acc, mf1, auc, ms]); cms[pol] = cm

# CSV 저장
import csv
csv_path = f"{OUT}/tta_results.csv"
with open(csv_path, "w", newline="") as f:
    w = csv.writer(f); w.writerow(["policy","accuracy","macro_f1","roc_auc","ms_per_image"]); w.writerows(rows)
print("Saved CSV:", csv_path)

# 혼동행렬도 정책별 저장(옵션)
for pol, cm in cms.items():
    plt.figure(figsize=(5.6,4.8))
    plt.imshow(cm, cmap='Blues'); plt.title(f'Confusion Matrix (Test, {pol})'); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    for i in range(len(classes)):
        for j in range(len(classes)):
            v=cm[i,j]; c='white' if v>cm.max()/2 else 'black'
            plt.text(j,i,str(v),ha='center',va='center',color=c,fontsize=9)
    plt.tight_layout(); plt.savefig(f"{OUT}/confmat_test_{pol}.png", dpi=220); plt.close()

print("Artifacts saved in:", OUT)
```

```python
# ============================================================
# CNN–ViT Hybrid (ResNet50 -> ViT)
# - 50 epochs (No EarlyStopping)
# - Warmup + Cosine LR schedule
# - MixUp/CutMix on-the-fly (train only)
# - Dropout + DropPath (Stochastic Depth)
# - Label smoothing 0.05
# - Macro-F1 & Accuracy checkpoints
# - Training curves + reports + confusion matrices saved
# ============================================================
import os, glob, random, shutil, numpy as np, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score

# 0) Mount & Paths
from google.colab import drive
drive.mount('/content/drive')

BASE   = '/content/drive/MyDrive/인공지능 튜플'
SRC    = f'{BASE}/OriginalDataset'      # 클래스 폴더 4개 (MildDemented, ModerateDemented, NonDemented, VeryMildDemented)
SPLIT  = '/content/dataset_hybrid'      # 분할 생성 위치
OUT    = f'{BASE}/proj_pr/out'; os.makedirs(OUT, exist_ok=True)

SEED = 123
random.seed(SEED); np.random.seed(SEED); tf.keras.utils.set_random_seed(SEED)

# -------------------
# 1) Split (class-wise 8:1:1). 매번 재생성(필요시 if 조건으로 유지 가능)
# -------------------
def ensure(p): os.makedirs(p, exist_ok=True)
def list_images(cls_dir):
    files = []
    for ext in ('*.png','*.jpg','*.jpeg','*.bmp'):
        files += glob.glob(os.path.join(cls_dir, ext))
    return files

if os.path.isdir(SPLIT): shutil.rmtree(SPLIT)
classes = sorted([d for d in os.listdir(SRC) if os.path.isdir(os.path.join(SRC,d))])
for sp in ('train','val','test'):
    for c in classes: ensure(os.path.join(SPLIT, sp, c))

summary = {}
for c in classes:
    files = list_images(os.path.join(SRC,c)); random.shuffle(files)
    n=len(files); n_val=n//10; n_test=n//10
    val_f = files[:n_val]; test_f = files[n_val:n_val+n_test]; train_f = files[n_val+n_test:]
    for fp in train_f: shutil.copy2(fp, os.path.join(SPLIT,'train',c,os.path.basename(fp)))
    for fp in val_f:   shutil.copy2(fp, os.path.join(SPLIT,'val',c,os.path.basename(fp)))
    for fp in test_f:  shutil.copy2(fp, os.path.join(SPLIT,'test',c,os.path.basename(fp)))
    summary[c]=(len(train_f),len(val_f),len(test_f))
print("[Split]")
for c,(tr,v,te) in summary.items():
    print(f"  {c:18s}  train={tr:5d}  val={v:4d}  test={te:4d}")

# -------------------
# 2) Datasets (balanced sampler for train)
# -------------------
IMG_SIZE=(288,288); BATCH=32
AUTOTUNE=tf.data.AUTOTUNE

def aug_strong(x):
    x = tf.image.random_flip_left_right(x)
    x = tf.image.random_brightness(x, 0.08)
    x = tf.image.random_contrast(x, 0.9, 1.1)
    x = tf.image.random_saturation(x, 0.9, 1.1)
    crop_scale = tf.random.uniform([], 0.8, 1.0)
    new_h = tf.cast(IMG_SIZE[0]*crop_scale, tf.int32)
    new_w = tf.cast(IMG_SIZE[1]*crop_scale, tf.int32)
    x = tf.image.resize(x, (new_h, new_w))
    x = tf.image.resize_with_pad(x, IMG_SIZE[0], IMG_SIZE[1])
    return x

def load_image(path, label, train=False):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, IMG_SIZE)
    if train: img = aug_strong(img)
    x = img/255.0
    return x, label

# per-class datasets with repeat + strong aug
per_class_files = {c: sorted(glob.glob(os.path.join(SPLIT,'train',c,'*'))) for c in classes}
per_class_ds, class_counts = [], []
for i,c in enumerate(classes):
    files = tf.constant(per_class_files[c])
    labels= tf.fill([tf.shape(files)[0]], i)
    ds = tf.data.Dataset.from_tensor_slices((files, labels)).shuffle(4096, seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(lambda p,l: load_image(p,l,train=True), num_parallel_calls=AUTOTUNE).repeat()
    per_class_ds.append(ds); class_counts.append(len(per_class_files[c]))

C = len(classes)
train_ds = tf.data.Dataset.sample_from_datasets(per_class_ds, weights=[1.0/C]*C, stop_on_empty_dataset=False)
train_ds = train_ds.map(lambda x,y: (x, tf.one_hot(y, C)), num_parallel_calls=AUTOTUNE)  # one-hot labels

# val/test (no aug)
val_raw  = tf.keras.utils.image_dataset_from_directory(os.path.join(SPLIT,'val') , image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)
test_raw = tf.keras.utils.image_dataset_from_directory(os.path.join(SPLIT,'test'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)

print("Class order (val):", val_raw.class_names)
assert val_raw.class_names==classes, "폴더/클래스 순서가 train과 다릅니다."

val_ds  = val_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, C))).prefetch(AUTOTUNE)
test_ds = test_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, C))).prefetch(AUTOTUNE)

# -------------------
# 3) MixUp/CutMix (Graph-safe)
# -------------------
def _sample_beta(alpha):
    a = tf.random.gamma([], alpha, dtype=tf.float32)
    b = tf.random.gamma([], alpha, dtype=tf.float32)
    return a / (a + b)

def _mixup(images, labels, lam):
    idx = tf.random.shuffle(tf.range(tf.shape(images)[0]))
    images2 = tf.gather(images, idx)
    labels2 = tf.gather(labels, idx)
    mixed  = lam*images + (1.0-lam)*images2
    labels = lam*labels + (1.0-lam)*labels2
    return mixed, labels

def _cutmix(images, labels, lam):
    B = tf.shape(images)[0]
    H = tf.shape(images)[1]; W = tf.shape(images)[2]; Cc = tf.shape(images)[3]
    idx = tf.random.shuffle(tf.range(B))
    images2 = tf.gather(images, idx)
    labels2 = tf.gather(labels, idx)

    r = tf.sqrt(1.0 - lam)
    cut_h = tf.cast(tf.cast(H, tf.float32)*r, tf.int32)
    cut_w = tf.cast(tf.cast(W, tf.float32)*r, tf.int32)
    cy = tf.random.uniform([], 0, H, dtype=tf.int32)
    cx = tf.random.uniform([], 0, W, dtype=tf.int32)
    y1 = tf.clip_by_value(cy - cut_h//2, 0, H)
    x1 = tf.clip_by_value(cx - cut_w//2, 0, W)
    y2 = tf.clip_by_value(y1 + cut_h, 0, H)
    x2 = tf.clip_by_value(x1 + cut_w, 0, W)

    Y = tf.range(H)[None, :, None]         # 1 x H x 1
    X = tf.range(W)[None, None, :]         # 1 x 1 x W
    rect = tf.cast((Y >= y1) & (Y < y2) & (X >= x1) & (X < x2), images.dtype)  # 1 x H x W
    rect = tf.expand_dims(rect, -1)        # 1 x H x W x 1
    mask = tf.tile(rect, [B, 1, 1, Cc])    # B x H x W x C

    mixed  = images*(1.0 - mask) + images2*mask
    box_area = tf.cast((y2-y1)*(x2-x1), tf.float32) / tf.cast(H*W, tf.float32)
    lam_eff  = 1.0 - box_area
    labels = lam_eff*labels + (1.0-lam_eff)*labels2
    return mixed, labels

@tf.function
def apply_mixup_cutmix(images, labels, alpha=0.2, p_cutmix=0.5):
    images = tf.cast(images, tf.float32)
    labels = tf.cast(labels, tf.float32)
    lam = _sample_beta(alpha)
    use_cutmix = tf.less(tf.random.uniform([], 0, 1), p_cutmix)
    mixed, out_labels = tf.cond(
        use_cutmix,
        lambda: _cutmix(images, labels, lam),
        lambda: _mixup(images, labels, lam)
    )
    return mixed, out_labels

# 배치 후 MixUp/CutMix 적용
train_ds = (train_ds
            .batch(BATCH)
            .map(lambda x,y: apply_mixup_cutmix(x, y, alpha=0.2, p_cutmix=0.5),
                 num_parallel_calls=AUTOTUNE)
            .prefetch(AUTOTUNE))

# -------------------
# 4) Model (DropPath 포함 Transformer Encoder)
# -------------------
class DropPath(layers.Layer):
    def __init__(self, rate=0.0, **kwargs):
        super().__init__(**kwargs); self.rate = float(rate)
    def call(self, x, training=False):
        if (not training) or self.rate <= 0.0: return x
        keep = 1.0 - self.rate
        shape = (tf.shape(x)[0],) + (1,)*(len(x.shape)-1)
        random_tensor = keep + tf.random.uniform(shape, dtype=x.dtype)
        binary_tensor = tf.floor(random_tensor)
        return (x / keep) * binary_tensor
    def get_config(self):
        c = super().get_config(); c.update({'rate': self.rate}); return c

class PositionalEmbedding2D(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.pos = None
    def build(self, input_shape):
        H, W, D = input_shape[1], input_shape[2], input_shape[3]
        if None in (H, W, D):
            raise ValueError(f"PositionalEmbedding2D needs static H,W,D, got {input_shape}")
        # ✅ Keras 3 호환: 모든 인자를 키워드로 전달
        self.pos = self.add_weight(
            name="pos2d",
            shape=(H, W, D),
            initializer=keras.initializers.RandomNormal(stddev=0.02),
            trainable=True,
        )
    def call(self, x): return x + self.pos
    def get_config(self):
        return super().get_config()

class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads=4, mlp_ratio=4.0, drop=0.2, droppath=0.1, **kwargs):
        super().__init__(**kwargs)
        self.n1 = layers.LayerNormalization(epsilon=1e-6)
        self.mha= layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim//num_heads, dropout=drop)
        self.dp1 = DropPath(rate=droppath)
        self.d1  = layers.Dropout(drop)
        self.n2  = layers.LayerNormalization(epsilon=1e-6)
        self.mlp = keras.Sequential([
            layers.Dense(int(embed_dim*mlp_ratio), activation=tf.nn.gelu),
            layers.Dropout(drop),
            layers.Dense(embed_dim),
            layers.Dropout(drop),
        ])
        self.dp2 = DropPath(rate=droppath)
    def call(self, x, training=False):
        h = self.mha(self.n1(x), self.n1(x), training=training)
        x = x + self.dp1(self.d1(h, training=training), training=training)
        h = self.mlp(self.n2(x), training=training)
        x = x + self.dp2(h, training=training)
        return x
    def get_config(self):
        return super().get_config()

def build_hybrid(input_shape, num_classes, embed_dim=256, vit_blocks=4, heads=4,
                 drop=0.2, droppath_max=0.1):
    inp = layers.Input(input_shape)
    x = keras.applications.resnet.preprocess_input(inp*255.0)
    backbone = keras.applications.ResNet50(include_top=False, weights='imagenet', name='resnet50')
    feat = backbone(x, training=True)

    tok = layers.Conv2D(embed_dim, 2, strides=2, padding='same', name='patch_conv')(feat)
    tok = layers.BatchNormalization(name='patch_bn')(tok)
    tok = layers.Activation('gelu', name='patch_gelu')(tok)
    tok = layers.Dropout(drop, name='patch_drop')(tok)

    tok = PositionalEmbedding2D(name='pos2d')(tok)
    H, W = tok.shape[1], tok.shape[2]
    z = layers.Reshape((H*W, embed_dim), name='tokens')(tok)
    z = layers.Dropout(drop, name='tok_drop')(z)

    dp_rates = np.linspace(0.0, droppath_max, vit_blocks)
    for i in range(vit_blocks):
        z = TransformerEncoder(embed_dim=embed_dim, num_heads=heads,
                               mlp_ratio=4.0, drop=drop, droppath=float(dp_rates[i]),
                               name=f'encoder_{i}')(z)

    z = layers.GlobalAveragePooling1D(name='gap')(z)
    z = layers.Dropout(0.5, name='head_drop1')(z)
    z = layers.Dense(512, activation='gelu', name='head_dense')(z)
    z = layers.Dropout(0.4, name='head_drop2')(z)
    out = layers.Dense(num_classes, activation='softmax', name='classifier')(z)

    model = keras.Model(inp, out, name='ResNet50_ViT_hybrid')
    for l in backbone.layers:
        if isinstance(l, layers.BatchNormalization): l.trainable = False
    return model

model = build_hybrid(IMG_SIZE+(3,), C, embed_dim=256, vit_blocks=4, heads=4,
                     drop=0.2, droppath_max=0.1)
model.summary()

# -------------------
# 5) Optim + Schedule (Warmup + Cosine), Compile
# -------------------
total_train_images = sum([n for n,_,_ in summary.values()])
steps_per_epoch = max(250, total_train_images//BATCH)   # 균형샘플이면 250~400 충분
TOTAL_EPOCHS = 50
TOTAL_STEPS  = steps_per_epoch * TOTAL_EPOCHS
WARMUP_STEPS = int(TOTAL_STEPS * 0.1)

class WarmupCosine(keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, base_lr, warmup_steps, total_steps):
        self.base_lr = base_lr; self.ws = float(warmup_steps); self.ts = float(total_steps)
    def __call__(self, step):
        step = tf.cast(step, tf.float32)
        lr_warm = self.base_lr * (step / tf.maximum(1.0, self.ws))
        progress = tf.clip_by_value((step - self.ws) / tf.maximum(1.0, self.ts - self.ws), 0., 1.)
        lr_cos = 0.5 * self.base_lr * (1 + tf.cos(np.pi * progress))
        return tf.where(step < self.ws, lr_warm, lr_cos)

lr_schedule = WarmupCosine(base_lr=5e-5, warmup_steps=WARMUP_STEPS, total_steps=TOTAL_STEPS)
opt = keras.optimizers.AdamW(learning_rate=lr_schedule, weight_decay=1e-4, clipnorm=1.0)

model.compile(
    optimizer=opt,
    loss=keras.losses.CategoricalCrossentropy(label_smoothing=0.05),
    metrics=['accuracy']
)

# -------------------
# 6) Callbacks (No EarlyStopping): macro-F1 & accuracy checkpoints
# -------------------
class MacroF1Checkpoint(keras.callbacks.Callback):
    def __init__(self, val_ds, path_model, path_weights):
        super().__init__(); self.val_ds = val_ds
        self.path_model = path_model; self.path_weights = path_weights; self.best = -1.0
    def on_epoch_end(self, epoch, logs=None):
        y_true, y_pred = [], []
        for x, y in self.val_ds:
            p = self.model.predict(x, verbose=0)
            y_true.extend(np.argmax(y.numpy(), axis=1))
            y_pred.extend(np.argmax(p, axis=1))
        f1 = f1_score(y_true, y_pred, average='macro')
        print(f"\n[val macroF1] {f1:.4f}")
        if f1 > self.best:
            self.best = f1
            #self.model.save(self.path_model, include_optimizer=False)           # .keras (전체 모델)
            self.model.save_weights(self.path_weights) # .weights.h5 (가중치만)
            print(f"[CKPT] Saved best (macroF1={self.best:.4f}) → {self.path_model} / {self.path_weights}")

ckpt_f1 = MacroF1Checkpoint(val_ds,
                            path_model = f'{OUT}/best_hybrid_by_f1.keras',
                            path_weights = f'{OUT}/best_hybrid_by_f1.weights.h5')

ckpt_acc = keras.callbacks.ModelCheckpoint(
    filepath=f'{OUT}/best_hybrid_by_acc.weights.h5',  # ← 확장자 수정(.weights.h5)
    monitor='val_accuracy', mode='max',
    save_best_only=True, save_weights_only=True, verbose=1
)

callbacks=[ckpt_f1, ckpt_acc]

# -------------------
# 7) Train (50 epochs)
# -------------------
hist = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=TOTAL_EPOCHS,
    steps_per_epoch=steps_per_epoch,
    callbacks=callbacks,
    verbose=1
)

# -------------------
# 8) Plots
# -------------------
epochs = range(1, len(hist.history['loss'])+1)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(epochs, hist.history['accuracy'], label='Train'); plt.plot(epochs, hist.history['val_accuracy'], label='Val')
plt.title('Accuracy'); plt.xlabel('Epoch'); plt.legend(); plt.grid(True)
plt.subplot(1,2,2)
plt.plot(epochs, hist.history['loss'], label='Train'); plt.plot(epochs, hist.history['val_loss'], label='Val')
plt.title('Loss'); plt.xlabel('Epoch'); plt.legend(); plt.grid(True)
plt.tight_layout(); plt.savefig(f'{OUT}/hybrid_training_curves.png', dpi=200); plt.show()

# -------------------
# 9) Reports (Val/Test, macro-F1)
# -------------------
def eval_report(mdl, ds, classes, tag):
    y_true, y_pred, probs = [], [], []
    for x, y in ds:
        p = mdl.predict(x, verbose=0)
        probs.append(p)
        y_true.extend(np.argmax(y.numpy(), axis=1))
        y_pred.extend(np.argmax(p, axis=1))
    y_true = np.array(y_true); y_pred = np.array(y_pred); probs = np.concatenate(probs,0)
    rep = classification_report(y_true, y_pred, target_names=classes, digits=4)
    print(f"\n[{tag} report]\n", rep)
    with open(f'{OUT}/hybrid_{tag}_report.txt','w') as f: f.write(rep)
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5.6,4.8))
    plt.imshow(cm, cmap='Blues'); plt.title(f'Confusion Matrix ({tag})'); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    for i in range(len(classes)):
        for j in range(len(classes)):
            v=cm[i,j]; c='white' if v>cm.max()/2 else 'black'
            plt.text(j,i,str(v),ha='center',va='center',color=c,fontsize=9)
    plt.tight_layout(); plt.savefig(f'{OUT}/hybrid_confmat_{tag}.png', dpi=220); plt.show()
    return f1_score(y_true, y_pred, average='macro')

classes = val_raw.class_names
val_f1  = eval_report(model, val_ds, classes, 'val_final')   # 마지막 epoch 모델
test_f1 = eval_report(model, test_ds, classes, 'test_final')

print(f"[Summary] Macro-F1  val={val_f1:.4f}  test={test_f1:.4f}")
print("Saved to:", OUT)
```

```python
# ============================================================
# CNN–ViT Hybrid (ResNet50 -> ViT) | Robust CKPT + Masked Rollout + TTA
# - CKPT 우선순위/형식 감지 로더: *.weights.h5 > .keras(zip) > .h5
# - Validation/Test 로더 (증강 X, [0,1] 스케일)
# - Attention Rollout (두상 마스크 내부 정규화 + 배경 알파=0)
# - gt|pred 타이틀 그리드 저장 + 진단 지표(brain-focus ratio)
# - TTA 정책별 성능 비교 (CSV/Confusion Matrix 저장)
# 저장 경로: /content/drive/MyDrive/인공지능 튜플/proj_pr/out/
# ============================================================
import os, glob, time, zipfile, csv
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, roc_auc_score

# -------------------
# 0) 경로/상수
# -------------------
from google.colab import drive
drive.mount('/content/drive')

BASE  = '/content/drive/MyDrive/인공지능 튜플'
SPLIT = '/content/dataset_hybrid'   # val/ , test/ 가 있는 경로
OUT   = f'{BASE}/proj_pr/out'
os.makedirs(OUT, exist_ok=True)

IMG_SIZE    = (288, 288)
BATCH       = 32
SEED        = 123
NUM_CLASSES = 4

tf.keras.utils.set_random_seed(SEED)

# -------------------
# 1) 모델 정의 (직렬화 가능)
# -------------------
class PositionalEmbedding2D(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.pos = None
    def build(self, input_shape):
        H, W, D = input_shape[1], input_shape[2], input_shape[3]
        if None in (H, W, D):
            raise ValueError(f"PositionalEmbedding2D needs static H,W,D, got {input_shape}")
        self.pos = self.add_weight(
            name="pos2d",
            shape=(H, W, D),
            initializer=keras.initializers.RandomNormal(stddev=0.02),
            trainable=True,
        )
    def call(self, x):  # x: (B,H,W,D)
        return x + self.pos
    def get_config(self):
        return super().get_config()

class TransformerEncoder(layers.Layer):
    def __init__(self, embed_dim, num_heads=4, mlp_ratio=4.0, drop=0.1, **kwargs):
        super().__init__(**kwargs)
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.mlp_ratio = mlp_ratio
        self.drop = drop
        self.n1 = layers.LayerNormalization(epsilon=1e-6)
        self.mha= layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim//num_heads, dropout=drop)
        self.d1 = layers.Dropout(drop)
        self.n2 = layers.LayerNormalization(epsilon=1e-6)
        self.mlp= keras.Sequential([
            layers.Dense(int(embed_dim*mlp_ratio), activation=tf.nn.gelu),
            layers.Dropout(drop),
            layers.Dense(embed_dim),
            layers.Dropout(drop),
        ])
    def call(self, x, training=False):
        h = self.mha(self.n1(x), self.n1(x), training=training)
        x = x + self.d1(h, training=training)
        x = x + self.mlp(self.n2(x), training=training)
        return x
    def get_config(self):
        cfg = super().get_config()
        cfg.update(dict(embed_dim=self.embed_dim, num_heads=self.num_heads,
                        mlp_ratio=self.mlp_ratio, drop=self.drop))
        return cfg

def build_hybrid(input_shape, num_classes, embed_dim=256, vit_blocks=4, heads=4, drop=0.1):
    inp = layers.Input(input_shape)
    x = keras.applications.resnet.preprocess_input(inp*255.0)
    backbone = keras.applications.ResNet50(include_top=False, weights='imagenet', name='resnet50')
    feat = backbone(x, training=False)

    tok = layers.Conv2D(embed_dim, 2, strides=2, padding='same', name='patch_embed_conv')(feat)
    tok = layers.BatchNormalization(name='patch_embed_bn')(tok)
    tok = layers.Activation('gelu', name='patch_embed_gelu')(tok)
    tok = layers.Dropout(drop, name='patch_embed_dropout')(tok)

    tok = PositionalEmbedding2D(name='pos2d')(tok)
    H, W = tok.shape[1], tok.shape[2]
    tokens = layers.Reshape((H*W, embed_dim), name='tokens')(tok)
    z = layers.Dropout(drop, name='tokens_dropout')(tokens)

    for i in range(vit_blocks):
        z = TransformerEncoder(embed_dim=embed_dim, num_heads=heads, mlp_ratio=4.0, drop=drop, name=f'encoder_{i}')(z)

    z = layers.GlobalAveragePooling1D(name='gap')(z)
    z = layers.Dropout(0.5, name='head_drop1')(z)
    z = layers.Dense(512, activation='gelu', name='head_dense')(z)
    z = layers.Dropout(0.4, name='head_drop2')(z)
    out = layers.Dense(num_classes, activation='softmax', name='classifier')(z)

    model = keras.Model(inp, out, name='ResNet50_ViT_hybrid')
    for l in backbone.layers:
        if isinstance(l, layers.BatchNormalization): l.trainable = False
    return model

# -------------------
# 2) CKPT 선택/로딩 (우선순위 + 형식 감지)
# -------------------
def pick_ckpt(out_dir):
    # 1순위: *.weights.h5  2순위: .keras(zip)  3순위: .h5
    w = sorted(glob.glob(os.path.join(out_dir, '*.weights.h5')), key=os.path.getmtime, reverse=True)
    k = sorted(glob.glob(os.path.join(out_dir, '*.keras')), key=os.path.getmtime, reverse=True)
    h = sorted([p for p in glob.glob(os.path.join(out_dir, '*.h5')) if not p.endswith('.weights.h5')], key=os.path.getmtime, reverse=True)
    cand = (w + k + h)
    print("[CKPT candidates]")
    for p in cand[:8]:
        sz = os.path.getsize(p) if os.path.exists(p) else -1
        print(" -", os.path.basename(p), f"({sz} bytes)")
    return cand[0] if cand else None

def is_valid_keras_zip(path):  # .keras가 진짜 zip인지 확인
    try:
        return zipfile.is_zipfile(path)
    except:
        return False

CKPT = pick_ckpt(OUT)
assert CKPT is not None, f"체크포인트 없음: {OUT}"

roll_model = None
if CKPT.endswith('.weights.h5'):
    print("[i] weights-only HDF5 → build + load_weights")
    tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
    tmp.load_weights(CKPT)
    roll_model = tmp
    print("[✓] load_weights:", os.path.basename(CKPT))

elif CKPT.endswith('.keras'):
    if is_valid_keras_zip(CKPT):
        print("[i] valid .keras zip → load_model")
        roll_model = keras.models.load_model(
            CKPT, compile=False,
            custom_objects={'PositionalEmbedding2D': PositionalEmbedding2D,
                            'TransformerEncoder': TransformerEncoder}
        )
        print("[✓] load_model(.keras):", os.path.basename(CKPT))
    else:
        print("[!] .keras가 zip 포맷이 아님 → weights로 시도")
        tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
        tmp.load_weights(CKPT)
        roll_model = tmp
        print("[✓] load_weights(.keras명):", os.path.basename(CKPT))
else:  # 기타 .h5
    print("[i] .h5 감지 → load_model 시도, 실패 시 weights로 재시도")
    try:
        roll_model = keras.models.load_model(
            CKPT, compile=False,
            custom_objects={'PositionalEmbedding2D': PositionalEmbedding2D,
                            'TransformerEncoder': TransformerEncoder}
        )
        print("[✓] load_model(.h5):", os.path.basename(CKPT))
    except Exception as e:
        print("[!] load_model 실패 → load_weights:", e)
        tmp = build_hybrid(IMG_SIZE+(3,), NUM_CLASSES, embed_dim=256, vit_blocks=4, heads=4, drop=0.1)
        tmp.load_weights(CKPT)
        roll_model = tmp
        print("[✓] load_weights(.h5):", os.path.basename(CKPT))

# -------------------
# 3) Val/Test 로더 (증강 X, [0,1])
# -------------------
val_raw = tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT, 'val'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)
test_raw = tf.keras.utils.image_dataset_from_directory(
    os.path.join(SPLIT, 'test'), image_size=IMG_SIZE, batch_size=BATCH, shuffle=False)

classes = val_raw.class_names
val_ds  = val_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, NUM_CLASSES))).prefetch(tf.data.AUTOTUNE)
test_ds = test_raw.map(lambda x,y: (x/255.0, tf.one_hot(y, NUM_CLASSES))).prefetch(tf.data.AUTOTUNE)
print("Classes:", classes)

# -------------------
# 4) Rollout 준비 (레이어 핸들/토큰 추출)
# -------------------
def get_layer_safe(model, name=None, type_=None, start_after=None):
    if name is not None:
        try:
            return model.get_layer(name)
        except:
            pass
    if type_ is not None:
        layers_ = model.layers
        if start_after is not None:
            layers_ = layers_[model.layers.index(start_after)+1:]
        for l in layers_:
            if isinstance(l, type_): return l
    raise ValueError(f"Layer not found: name={name}, type={type_}")

L = roll_model.layers
resnet = get_layer_safe(roll_model, name='resnet50')
conv_pe = get_layer_safe(roll_model, name='patch_embed_conv', type_=layers.Conv2D, start_after=resnet)
bn_pe   = get_layer_safe(roll_model, name='patch_embed_bn',   type_=layers.BatchNormalization)
gelu_pe = get_layer_safe(roll_model, name='patch_embed_gelu')
drop0   = get_layer_safe(roll_model, name='patch_embed_dropout', type_=layers.Dropout)
pos2d   = get_layer_safe(roll_model, name='pos2d', type_=PositionalEmbedding2D)
reshape = get_layer_safe(roll_model, name='tokens', type_=layers.Reshape)
drop_tk = get_layer_safe(roll_model, name='tokens_dropout', type_=layers.Dropout)
encoders = [l for l in L if isinstance(l, TransformerEncoder)]
assert encoders, "TransformerEncoder 레이어를 찾지 못했습니다."

@tf.function(jit_compile=False)
def tokens_from_image(x):  # x: [0,1]
    x = keras.applications.resnet.preprocess_input(x*255.0)
    f = resnet(x, training=False)
    t = conv_pe(f, training=False)
    t = bn_pe(t, training=False)
    t = gelu_pe(t)
    t = drop0(t, training=False)
    t = pos2d(t, training=False)
    H = tf.shape(t)[1]; W = tf.shape(t)[2]
    z = reshape(t); z = drop_tk(z, training=False)  # (B,T,D)
    return z, H, W

# --- 안전 assert ---
def _assert_row_stochastic(A, atol=1e-4):  # A: (B,Tq,Tk)
    row_sum = tf.reduce_sum(A, axis=-1)
    tf.debugging.assert_near(row_sum, tf.ones_like(row_sum), atol=atol,
                             message="Attention rows must sum to 1 (row-stochastic).")

def _assert_grid_tokens(H, W, T):
    tf.debugging.assert_equal(tf.math.multiply(H, W), T,
                              message="Grid H*W must equal #tokens (reshape mismatch).")

# -------------------
# 5) 두상 마스크 + 마스크 내부 정규화 롤아웃
# -------------------
def brain_mask_smart(x, rel=0.20, min_thr=0.05, morph=2):
    """x: [0,1] NHWC → 상대 임계 + 모폴로지 닫기로 두상 마스크"""
    g = tf.image.rgb_to_grayscale(x)                         # (B,H,W,1)
    vmax = tf.reduce_max(g, axis=[1,2,3], keepdims=True)
    thr  = tf.maximum(min_thr, rel * vmax)
    m = tf.cast(g > thr, tf.float32)                         # 0/1

    k = 3
    for _ in range(morph):                                   # dilate
        m = tf.nn.max_pool2d(m, ksize=k, strides=1, padding='SAME')
    for _ in range(morph):                                   # erode
        m = 1.0 - tf.nn.max_pool2d(1.0 - m, ksize=k, strides=1, padding='SAME')
    return m                                                 # (B,H,W,1)

def compute_rollout_map_masked(
    images, head_fuse='max', use_residual=True,
    mask_rel=0.20, mask_min_thr=0.05, gamma=0.75
):
    images_tf = tf.convert_to_tensor(images, dtype=tf.float32) if isinstance(images, np.ndarray) else images
    z, H, W = tokens_from_image(images_tf)
    B = tf.shape(z)[0]; T = tf.shape(z)[1]
    _assert_grid_tokens(H, W, T)

    # Rollout 누적
    R = tf.eye(T, batch_shape=[B])
    cur = z
    for enc in encoders:
        q = enc.n1(cur)
        attn_out, scores = enc.mha(q, q, return_attention_scores=True, training=False)  # (B,H,T,T)
        A = tf.reduce_max(scores, axis=1) if head_fuse=='max' else tf.reduce_mean(scores, axis=1)
        _assert_row_stochastic(A)
        if use_residual:
            A = A + tf.eye(T, batch_shape=[B])
        A = A / tf.reduce_sum(A, axis=-1, keepdims=True)
        R = tf.linalg.matmul(R, A)

        cur = cur + enc.d1(attn_out, training=False)
        h = enc.n2(cur); h = enc.mlp(h, training=False)
        cur = cur + h

    heat = tf.reduce_mean(R, axis=1)                          # (B,T)
    heat = tf.reshape(heat, (-1, H, W, 1))
    heat = tf.image.resize(heat, tf.shape(images_tf)[1:3], method='bilinear')

    # 마스크 내부 기준 정규화
    m = brain_mask_smart(images_tf, rel=mask_rel, min_thr=mask_min_thr)      # (B,H,W,1)
    heat = heat * m

    very_neg = tf.constant(-1e9, heat.dtype)
    very_pos = tf.constant(1e9,  heat.dtype)
    hmax = tf.reduce_max(tf.where(m>0, heat, very_neg), axis=[1,2,3], keepdims=True)
    hmin = tf.reduce_min(tf.where(m>0, heat, very_pos), axis=[1,2,3], keepdims=True)
    heat = tf.clip_by_value((heat - hmin) / (hmax - hmin + 1e-8), 0., 1.)
    heat = tf.pow(heat, gamma)                                 # 스팟 강조

    return heat.numpy()[..., 0], m.numpy()[..., 0]             # (B,H,W), (B,H,W)

def save_rollout_grid_masked(ds, mdl, classes, save_path, n=12, cols=6,
                             head_fuse='max', mask_rel=0.20, mask_min_thr=0.05, gamma=0.75):
    # 샘플 뽑기
    xs, ys = [], []
    for x, y in ds.unbatch().take(n):
        xs.append(x.numpy()); ys.append(np.argmax(y.numpy()))
    xs = np.stack(xs, axis=0); ys = np.array(ys)

    # 예측
    probs = mdl.predict(xs, verbose=0)
    preds = np.argmax(probs, axis=1)

    # 히트맵 + 마스크
    heat, mask = compute_rollout_map_masked(xs, head_fuse=head_fuse,
                                            mask_rel=mask_rel, mask_min_thr=mask_min_thr, gamma=gamma)

    # 그리드 렌더링 (배경 알파=0)
    rows = int(np.ceil(n/cols))
    fig, axes = plt.subplots(rows, cols, figsize=(3.4*cols, 3.6*rows))
    axes = np.array(axes).reshape(rows, cols)
    k = 0
    for r in range(rows):
        for c in range(cols):
            ax = axes[r, c]; ax.axis('off')
            if k < n:
                ax.imshow(xs[k])
                ax.imshow(heat[k], cmap='jet', alpha=0.45*mask[k])  # ★ 배경 투명
                gt = classes[ys[k]]; pd = classes[preds[k]]
                ax.set_title(f"gt={gt} | pred={pd}", fontsize=11)
                k += 1
    plt.tight_layout()
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    plt.savefig(save_path, dpi=200); plt.close()

def rollout_diagnostics(ds, mdl, head_fuse='max', n=48,
                        mask_rel=0.20, mask_min_thr=0.05, gamma=0.75):
    xs = np.stack([x.numpy() for x,_ in ds.unbatch().take(n)], axis=0)
    heat, mask = compute_rollout_map_masked(xs, head_fuse=head_fuse,
                                            mask_rel=mask_rel, mask_min_thr=mask_min_thr, gamma=gamma)
    inside = float((heat*mask).sum())
    total  = float(heat.sum() + 1e-8)
    brain_focus = inside/total
    # 토큰 그리드 진단
    z,H,W = tokens_from_image(tf.convert_to_tensor(xs, tf.float32))
    T = int(z.shape[1])
    print(f"[diag] grid HxW={int(H.numpy())}x{int(W.numpy())}, T={T}, ok={int(H.numpy())*int(W.numpy())==T}")
    print(f"[diag] brain-focus ratio={brain_focus:.3f} (권장 ≥ 0.90)")
    return brain_focus

# -------------------
# 6) 예시: 그리드/진단 실행
# -------------------
grid_val = f"{OUT}/rollout/val_rollout_grid_masked.png"
save_rollout_grid_masked(val_ds, roll_model, classes, grid_val,
                         n=12, cols=6, head_fuse='max', mask_rel=0.20, mask_min_thr=0.05, gamma=0.75)
print("Saved:", grid_val)

# 원하면 test도
grid_test = f"{OUT}/rollout/test_rollout_grid_masked.png"
save_rollout_grid_masked(test_ds, roll_model, classes, grid_test,
                         n=12, cols=6, head_fuse='max', mask_rel=0.20, mask_min_thr=0.05, gamma=0.75)
print("Saved:", grid_test)

# 수치 진단
rollout_diagnostics(val_ds, roll_model, head_fuse='max', n=48,
                    mask_rel=0.20, mask_min_thr=0.05, gamma=0.75)

# -------------------
# 7) TTA 성능 비교 (Test)
# -------------------
def five_crops(x, pad=16):
    H, W = x.shape[1], x.shape[2]
    x_pad = tf.pad(x, [[0,0],[pad,pad],[pad,pad],[0,0]], mode='REFLECT')
    crops = []
    crops.append(tf.image.crop_to_bounding_box(x_pad, 0, 0,     H, W))       # TL
    crops.append(tf.image.crop_to_bounding_box(x_pad, 0, 2*pad, H, W))       # TR
    crops.append(tf.image.crop_to_bounding_box(x_pad, 2*pad, 0, H, W))       # BL
    crops.append(tf.image.crop_to_bounding_box(x_pad, 2*pad, 2*pad, H, W))   # BR
    crops.append(tf.image.crop_to_bounding_box(x_pad, pad, pad, H, W))       # Center
    return crops

def predict_tta(batch_x, policy, mdl):
    if policy == 'none':
        views = [batch_x]
    elif policy == 'hflip':
        views = [batch_x, tf.image.flip_left_right(batch_x)]
    elif policy == '5crop':
        views = five_crops(batch_x)
    elif policy == 'hflip+5crop':
        base = five_crops(batch_x); views = base + [tf.image.flip_left_right(c) for c in base]
    else:
        raise ValueError("unknown TTA policy")
    probs = [mdl.predict(v, verbose=0) for v in views]
    return np.stack(probs, axis=0).mean(axis=0)

def eval_ds_tta(ds, classes, policy, mdl):
    y_true, y_pred, probs = [], [], []
    n_img = 0; t0 = time.time()
    for x, y in ds:
        p = predict_tta(x, policy, mdl)
        probs.append(p)
        y_true.extend(np.argmax(y.numpy(), axis=1))
        y_pred.extend(np.argmax(p, axis=1))
        n_img += x.shape[0]
    dt = time.time() - t0

    probs = np.concatenate(probs, 0)
    y_true = np.array(y_true); y_pred = np.array(y_pred)
    acc = accuracy_score(y_true, y_pred)
    mf1 = f1_score(y_true, y_pred, average='macro')
    auc = roc_auc_score(y_true, probs, multi_class='ovr', average='macro')
    ms_img = dt / n_img * 1000.0
    cm = confusion_matrix(y_true, y_pred)

    print(f"[TTA:{policy:11s}] acc={acc:.4f}  macroF1={mf1:.4f}  ROC-AUC={auc:.4f}  {ms_img:.1f} ms/img")
    return acc, mf1, auc, ms_img, cm

policies = ['none', 'hflip', '5crop', 'hflip+5crop']
rows = []; cms = {}
for pol in policies:
    acc, mf1, auc, ms, cm = eval_ds_tta(test_ds, classes, pol, roll_model)
    rows.append([pol, acc, mf1, auc, ms]); cms[pol] = cm

csv_path = f"{OUT}/tta_results.csv"
with open(csv_path, "w", newline="") as f:
    w = csv.writer(f); w.writerow(["policy","accuracy","macro_f1","roc_auc","ms_per_image"]); w.writerows(rows)
print("Saved CSV:", csv_path)

for pol, cm in cms.items():
    plt.figure(figsize=(5.6,4.8))
    plt.imshow(cm, cmap='Blues'); plt.title(f'Confusion Matrix (Test, {pol})'); plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes, rotation=45, ha='right'); plt.yticks(ticks, classes)
    for i in range(len(classes)):
        for j in range(len(classes)):
            v=cm[i,j]; c='white' if v>cm.max()/2 else 'black'
            plt.text(j,i,str(v),ha='center',va='center',color=c,fontsize=9)
    plt.tight_layout(); plt.savefig(f"{OUT}/confmat_test_{pol}.png", dpi=220); plt.close()

print("Artifacts saved in:", OUT)
```
